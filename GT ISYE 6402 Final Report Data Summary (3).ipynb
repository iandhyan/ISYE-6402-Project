This file contains the Data Summary table code of all the Datasets we used in our report. All the datasets can be found in the Additional Data Folder in the Zip File.

```python
import pandas as pd

BASE_PATH = "/Users/ranjaniravib/Documents/GT ISYE 6402/"

def load_series(filename):
    path = BASE_PATH + filename
    df = pd.read_csv(path)
    date_col = [c for c in df.columns if "DATE" in c.upper()][0]
    val_col = [c for c in df.columns if c != date_col][0]
    df[date_col] = pd.to_datetime(df[date_col])
    df = df.sort_values(date_col)
    df = df[[date_col, val_col]].dropna()
    df = df.set_index(date_col)
    return df[val_col]

paths = {
    "TOTALSA": "TOTALSA (1).csv",
    "AMVPNO": "AMVPNO (1).csv",
    "AISRSA": "AISRSA (1).csv",
    "GASREGCOVM": "GASREGCOVM (2).csv",
    "CPIAUCSL": "CPIAUCSL (1).csv",
    "FEDFUNDS": "FEDFUNDS.csv",
    "DAUTOSA": "DAUTOSA.csv",
    "FAUTOSA": "FAUTOSA.csv"
}

rows = []

for name, fname in paths.items():
    s = load_series(fname)
    rows.append({
        "Series": name,
        "Start": s.index.min().date(),
        "End": s.index.max().date(),
        "Obs": len(s),
        "Mean": round(s.mean(), 4),
        "Std Dev": round(s.std(), 4),
        "Min": round(s.min(), 4),
        "Max": round(s.max(), 4)
    })

summary_df = pd.DataFrame(rows)
summary_df
```
